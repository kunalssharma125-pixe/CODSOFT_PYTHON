```python
import random
import time

choices = ["rock", "paper", "scissors"]

user_score = 0
computer_score = 0
tie = 0
total_match = 0

history = []

messages = [
    "Awesome Move!",
    "Nice Choice!",
    "Better Luck Next Time!",
    "Excellent!",
    "Keep Playing!",
    "That was Close!",
    "Fantastic!"
]

print("=" * 50)
print("      ROCK PAPER SCISSORS GAME")
print("=" * 50)

while True:

    print("\n")
    print("=" * 50)
    print("1. Play Game")
    print("2. View Rules")
    print("3. View Score")
    print("4. Reset Score")
    print("5. Match History")
    print("6. Exit")

    menu = input("\nEnter your choice: ")

    # ---------------- PLAY GAME ----------------

    if menu == "1":

        print("\nChoose Your Move")
        print("1. Rock")
        print("2. Paper")
        print("3. Scissors")

        user = input("Enter Choice: ")

        if user == "1":
            user = "rock"

        elif user == "2":
            user = "paper"

        elif user == "3":
            user = "scissors"

        elif user.lower() in choices:
            user = user.lower()

        else:
            print("Invalid Choice.")
            continue

        print("\nRock...")
        time.sleep(1)

        print("Paper...")
        time.sleep(1)

        print("Scissors...")
        time.sleep(1)

        print("GO!!\n")

        computer = random.choice(choices)

        print("Your Choice     :", user.title())
        print("Computer Choice :", computer.title())

        total_match += 1

        if user == computer:

            print("\nResult : Match Tied")

            tie += 1

            winner = "Tie"

        elif (user == "rock" and computer == "scissors") or \
             (user == "paper" and computer == "rock") or \
             (user == "scissors" and computer == "paper"):

            print("\nCongratulations! You Win.")

            user_score += 1

            winner = "You"

        else:

            print("\nComputer Wins.")

            computer_score += 1

            winner = "Computer"

        history.append(
            "Match " + str(total_match) +
            " | You : " + user.title() +
            " | Computer : " + computer.title() +
            " | Winner : " + winner
        )

        print("\n" + random.choice(messages))

        again = input("\nPlay Again? (y/n): ")

        if again.lower() != "y":
            continue

    # ---------------- RULES ----------------

    elif menu == "2":

        print("\n========== RULES ==========")
        print("Rock beats Scissors")
        print("Paper beats Rock")
        print("Scissors beats Paper")

    # ---------------- SCORE ----------------

    elif menu == "3":

        print("\n========== SCORE ==========")

        print("You      :", user_score)
        print("Computer :", computer_score)
        print("Tie      :", tie)
        print("Matches  :", total_match)

        if total_match > 0:
            percentage = (user_score / total_match) * 100
            print("Win Percentage :", round(percentage,2), "%")
        else:
            print("Win Percentage : 0%")

    # ---------------- RESET ----------------

    elif menu == "4":

        ans = input("Reset all scores? (y/n): ")

        if ans.lower() == "y":

            user_score = 0
            computer_score = 0
            tie = 0
            total_match = 0
            history.clear()

            print("Scores Reset Successfully.")

        else:

            print("Reset Cancelled.")

    # ---------------- HISTORY ----------------

    elif menu == "5":

        if len(history) == 0:

            print("\nNo Match History Found.")

        else:

            print("\n========== MATCH HISTORY ==========")

            for item in history:
                print(item)

    # ---------------- EXIT ----------------

    elif menu == "6":

        print("\n")
        print("=" * 50)
        print("             GAME OVER")
        print("=" * 50)

        print("Total Matches :", total_match)
        print("You Won       :", user_score)
        print("Computer Won  :", computer_score)
        print("Tie Matches   :", tie)

        if total_match > 0:
            percentage = (user_score / total_match) * 100
            print("Win Percentage:", round(percentage,2), "%")

        if user_score > computer_score:

            print("\nOverall Winner : YOU")

        elif computer_score > user_score:

            print("\nOverall Winner : COMPUTER")

        else:

            print("\nOverall Result : DRAW")

        print("\nThank You For Playing!")
        break

    else:

        print("Invalid Choice.")
```
