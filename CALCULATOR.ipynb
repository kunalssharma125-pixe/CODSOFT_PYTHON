```python
import math

history = []

while True:

    print("\n" + "=" * 45)
    print("         PYTHON CALCULATOR")
    print("=" * 45)
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Modulus")
    print("6. Power")
    print("7. Floor Division")
    print("8. Square Root")
    print("9. Percentage")
    print("10. Maximum")
    print("11. Minimum")
    print("12. View History")
    print("13. Clear History")
    print("14. Exit")

    choice = input("\nEnter your choice: ")

    if choice == "14":
        print("\nThank You For Using Calculator.")
        break

    elif choice == "12":

        if len(history) == 0:
            print("\nNo History Available.")
        else:
            print("\n------ Calculation History ------")
            for item in history:
                print(item)

    elif choice == "13":

        history.clear()
        print("\nHistory Cleared Successfully.")

    elif choice == "8":

        num = float(input("Enter Number: "))

        if num < 0:
            print("Square Root of negative number is not possible.")
        else:
            result = math.sqrt(num)
            print("Answer =", result)
            history.append(f"√{num} = {result}")

    elif choice == "9":

        total = float(input("Enter Total Value: "))
        value = float(input("Enter Obtained Value: "))

        if total == 0:
            print("Total cannot be zero.")
        else:
            result = (value / total) * 100
            print("Percentage =", result, "%")
            history.append(f"{value}/{total} = {result}%")

    elif choice in ["1","2","3","4","5","6","7","10","11"]:

        num1 = float(input("Enter First Number: "))
        num2 = float(input("Enter Second Number: "))

        if choice == "1":
            result = num1 + num2
            print("Answer =", result)
            history.append(f"{num1} + {num2} = {result}")

        elif choice == "2":
            result = num1 - num2
            print("Answer =", result)
            history.append(f"{num1} - {num2} = {result}")

        elif choice == "3":
            result = num1 * num2
            print("Answer =", result)
            history.append(f"{num1} × {num2} = {result}")

        elif choice == "4":

            if num2 == 0:
                print("Cannot Divide By Zero.")
            else:
                result = num1 / num2
                print("Answer =", result)
                history.append(f"{num1} / {num2} = {result}")

        elif choice == "5":

            if num2 == 0:
                print("Cannot Perform Modulus By Zero.")
            else:
                result = num1 % num2
                print("Answer =", result)
                history.append(f"{num1} % {num2} = {result}")

        elif choice == "6":

            result = num1 ** num2
            print("Answer =", result)
            history.append(f"{num1}^{num2} = {result}")

        elif choice == "7":

            if num2 == 0:
                print("Cannot Divide By Zero.")
            else:
                result = num1 // num2
                print("Answer =", result)
                history.append(f"{num1}//{num2} = {result}")

        elif choice == "10":

            if num1 > num2:
                print("Maximum =", num1)
                history.append(f"Max({num1},{num2}) = {num1}")
            else:
                print("Maximum =", num2)
                history.append(f"Max({num1},{num2}) = {num2}")

        elif choice == "11":

            if num1 < num2:
                print("Minimum =", num1)
                history.append(f"Min({num1},{num2}) = {num1}")
            else:
                print("Minimum =", num2)
                history.append(f"Min({num1},{num2}) = {num2}")

    else:
        print("Invalid Choice.")
```
