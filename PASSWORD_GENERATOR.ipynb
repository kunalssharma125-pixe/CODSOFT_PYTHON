```python
import random
import string

history = []

tips = [
    "Use a different password for every account.",
    "Never share your password with anyone.",
    "Use at least 12 characters for better security.",
    "Change your passwords regularly.",
    "Include letters, numbers and symbols."
]

while True:

    print("\n" + "=" * 50)
    print("         ADVANCED PASSWORD GENERATOR")
    print("=" * 50)
    print("1. Generate Password")
    print("2. Generate 5 Passwords")
    print("3. Check Password Strength")
    print("4. View Password History")
    print("5. Save Password History")
    print("6. Clear Password History")
    print("7. Exit")

    choice = input("\nEnter your choice: ")

    # ------------------------------------
    # Generate Password
    # ------------------------------------

    if choice == "1" or choice == "2":

        try:
            length = int(input("\nEnter Password Length (4-50): "))

            if length < 4 or length > 50:
                print("Password length should be between 4 and 50.")
                continue

        except:
            print("Please enter a valid number.")
            continue

        print("\nSelect Character Types")

        upper = input("Include Uppercase Letters? (y/n): ")
        lower = input("Include Lowercase Letters? (y/n): ")
        number = input("Include Numbers? (y/n): ")
        symbol = input("Include Special Characters? (y/n): ")
        avoid = input("Avoid Similar Characters (O,0,I,l)? (y/n): ")

        characters = ""

        if upper.lower() == "y":
            characters += string.ascii_uppercase

        if lower.lower() == "y":
            characters += string.ascii_lowercase

        if number.lower() == "y":
            characters += string.digits

        if symbol.lower() == "y":
            characters += string.punctuation

        if characters == "":
            print("\nPlease select at least one character type.")
            continue

        if avoid.lower() == "y":

            characters = characters.replace("O", "")
            characters = characters.replace("0", "")
            characters = characters.replace("I", "")
            characters = characters.replace("l", "")

        if choice == "1":
            total = 1
        else:
            total = 5

        count = 1

        while count <= total:

            password_list = []

            if upper.lower() == "y":
                password_list.append(random.choice(string.ascii_uppercase.replace("O", "").replace("I", "")) if avoid.lower() == "y"
                                     else random.choice(string.ascii_uppercase))

            if lower.lower() == "y":
                password_list.append(random.choice(string.ascii_lowercase.replace("l", "")) if avoid.lower() == "y"
                                     else random.choice(string.ascii_lowercase))

            if number.lower() == "y":
                password_list.append(random.choice(string.digits.replace("0", "")) if avoid.lower() == "y"
                                     else random.choice(string.digits))

            if symbol.lower() == "y":
                password_list.append(random.choice(string.punctuation))

            while len(password_list) < length:
                password_list.append(random.choice(characters))

            random.shuffle(password_list)

            password = ""

            for ch in password_list:
                password = password + ch

            history.append(password)

            print("\nPassword", count)
            print("----------------------------")
            print(password)

            print("\nPassword Details")
            print("Length     :", len(password))
            print("Uppercase  :", any(c.isupper() for c in password))
            print("Lowercase  :", any(c.islower() for c in password))
            print("Numbers    :", any(c.isdigit() for c in password))
            print("Symbols    :", any(not c.isalnum() for c in password))

            print("\nSecurity Tip:")
            print(random.choice(tips))

            count += 1

    # ------------------------------------
    # Password Strength
    # ------------------------------------

    elif choice == "3":

        password = input("\nEnter Password: ")

        score = 0

        if len(password) >= 8:
            score += 1

        if any(c.isupper() for c in password):
            score += 1

        if any(c.islower() for c in password):
            score += 1

        if any(c.isdigit() for c in password):
            score += 1

        if any(not c.isalnum() for c in password):
            score += 1

        print("\nPassword Strength")

        if score <= 2:
            print("⭐ Weak Password")

        elif score <= 4:
            print("⭐⭐⭐ Medium Password")

        else:
            print("⭐⭐⭐⭐⭐ Strong Password")

    # ------------------------------------
    # View History
    # ------------------------------------

    elif choice == "4":

        if len(history) == 0:
            print("\nNo Password History Found.")

        else:

            print("\nGenerated Password History")
            print("----------------------------")

            print("Total Passwords :", len(history))

            count = 1

            for password in history:
                print(str(count) + ".", password)
                count += 1

    # ------------------------------------
    # Save History
    # ------------------------------------

    elif choice == "5":

        file = open("password_history.txt", "w")

        count = 1

        for password in history:

            file.write(str(count) + ". " + password + "\n")

            count += 1

        file.close()

        print("\nPassword History Saved Successfully.")

    # ------------------------------------
    # Clear History
    # ------------------------------------

    elif choice == "6":

        if len(history) == 0:
            print("\nHistory is already empty.")

        else:

            answer = input("Are you sure? (y/n): ")

            if answer.lower() == "y":
                history.clear()
                print("Password History Cleared.")
            else:
                print("Operation Cancelled.")

    # ------------------------------------
    # Exit
    # ------------------------------------

    elif choice == "7":

        print("\nThank You For Using Password Generator.")
        break

    else:

        print("\nInvalid Choice. Please Try Again.")
```
